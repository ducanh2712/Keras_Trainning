```python
import matplotlib.pyplot as plt
import numpy as np
import os
import PIL
import tensorflow as tf
from tensorflow import keras
from keras import layers
from keras.layers import Dense, Flatten
from keras.models import Sequential
from keras.optimizers import Adam
```

```python
IMAGE_SHAPE = (224, 224)


DATA_DIR = 'D:/OneDrive - Hanoi University of Science and Technology/BackUp/EDABK/AI/Data_AI_EDABK/TuTheNgu/EDABK_IR_23CLASS_LOSO'
TRAINING_DATA_DIR = []
VALID_DATA_DIR = []

for fold in os.listdir(DATA_DIR):
    TRAINING_DATA_DIR.append(DATA_DIR + '/' + fold + '/' + 'train_aug')
    VALID_DATA_DIR.append(DATA_DIR + '/' + fold + '/' + 'test')
print(TRAINING_DATA_DIR)
```

```python
train_generator = []
valid_generator = []

img_height,img_width=180,180
batch_size=32

print(range(len(os.listdir(DATA_DIR))))
for i in range(len(os.listdir(DATA_DIR))):
    train_generator.append(tf.keras.preprocessing.image_dataset_from_directory(
        TRAINING_DATA_DIR[i],
        seed=123,
        label_mode='categorical',
        image_size=(img_height, img_width),
        batch_size=batch_size
    ))
    valid_generator.append(tf.keras.preprocessing.image_dataset_from_directory(
        VALID_DATA_DIR[i],
        seed=123,
        label_mode='categorical',
        image_size=(img_height, img_width),
        batch_size=batch_size
    ))
```

```python
resnet_model = Sequential()

pretrained_model= tf.keras.applications.ResNet50(include_top=False,
                   input_shape=(180,180,3),
                   pooling='avg',classes=23,
                   weights='imagenet')
for layer in pretrained_model.layers:
        layer.trainable=False

resnet_model.add(pretrained_model)
```

```python
resnet_model.add(Flatten())
resnet_model.add(Dense(512, activation='relu'))
resnet_model.add(Dense(23, activation='softmax'))

resnet_model.summary()
```

```python
resnet_model.compile(optimizer=Adam(lr=0.001),loss='categorical_crossentropy',metrics=['accuracy'])
```

```python
history = []

for i in range(len(os.listdir(DATA_DIR))):
    history.append(resnet_model.fit(train_generator[i], validation_data=valid_generator[i], epochs=10))
    
# Epoch 1/10
# 597/597 [==============================] - 112s 174ms/step - loss: 1.3847 - accuracy: 0.5500 - val_loss: 0.7751 - val_accuracy: 0.7599
# Epoch 2/10
# 597/597 [==============================] - 99s 166ms/step - loss: 0.6162 - accuracy: 0.7886 - val_loss: 0.4735 - val_accuracy: 0.8375
# Epoch 3/10
# 597/597 [==============================] - 100s 167ms/step - loss: 0.4171 - accuracy: 0.8574 - val_loss: 0.4101 - val_accuracy: 0.8458
# Epoch 4/10
# 597/597 [==============================] - 99s 165ms/step - loss: 0.2865 - accuracy: 0.9063 - val_loss: 0.3349 - val_accuracy: 0.8840
# Epoch 5/10
# 597/597 [==============================] - 99s 165ms/step - loss: 0.2398 - accuracy: 0.9191 - val_loss: 0.2516 - val_accuracy: 0.9158
# Epoch 6/10
# 597/597 [==============================] - 99s 166ms/step - loss: 0.1864 - accuracy: 0.9375 - val_loss: 0.2464 - val_accuracy: 0.9208
# Epoch 7/10
# 597/597 [==============================] - 100s 167ms/step - loss: 0.1411 - accuracy: 0.9543 - val_loss: 0.2394 - val_accuracy: 0.9129
# Epoch 8/10
# 597/597 [==============================] - 100s 167ms/step - loss: 0.1325 - accuracy: 0.9545 - val_loss: 0.3081 - val_accuracy: 0.8957
# Epoch 9/10
# 597/597 [==============================] - 99s 166ms/step - loss: 0.1145 - accuracy: 0.9616 - val_loss: 0.1935 - val_accuracy: 0.9367
# Epoch 10/10
# 597/597 [==============================] - 99s 165ms/step - loss: 0.1004 - accuracy: 0.9654 - val_loss: 0.2417 - val_accuracy: 0.9158
# Epoch 1/10
# 597/597 [==============================] - 103s 171ms/step - loss: 0.1663 - accuracy: 0.9445 - val_loss: 0.1075 - val_accuracy: 0.9568
# Epoch 2/10
# 597/597 [==============================] - 100s 167ms/step - loss: 0.1010 - accuracy: 0.9677 - val_loss: 0.1333 - val_accuracy: 0.9548
# Epoch 3/10
# ...
# Epoch 9/10
# 597/597 [==============================] - 107s 179ms/step - loss: 1.6218e-04 - accuracy: 1.0000 - val_loss: 8.9172e-04 - val_accuracy: 0.9996
# Epoch 10/10
# 597/597 [==============================] - 108s 180ms/step - loss: 0.0826 - accuracy: 0.9810 - val_loss: 0.2106 - val_accuracy: 0.9296
```
