```python
# import the needed libraries
from tensorflow.keras.models import load_model
from PIL import Image
from matplotlib import pyplot as plt
import numpy as np
import cv2

# read the input image using Pillow (you can use another library, e.g., OpenCV)
img1= Image.open("D:/OneDrive - Hanoi University of Science and Technology/BackUp/EDABK/AI/Data_AI_EDABK/TuTheNgu/EDABK_IR_23CLASS_LOSO/S1/test/1/Frame0.png")
img2= Image.open("D:/OneDrive - Hanoi University of Science and Technology/BackUp/EDABK/AI/Data_AI_EDABK/TuTheNgu/EDABK_IR_23CLASS_LOSO/S1/test/9/Frame1062.png")

# convert to ndarray numpy
img1 = np.asarray(img1)
img2 = np.asarray(img2)

img1 = cv2.resize(img1,(224,224))
img2 = cv2.resize(img2,(224,224))

# load the trained model
model = load_model('./model/modelLOSO_Test.h5')

# predict the input image using the loaded model
pred1 = model.predict((img1/255).reshape(1,224,224,3))
pred2 = model.predict((img2/255).reshape(1,224,224,3))

#plot the prediction result
plt.figure('img1')
plt.imshow(img1)
plt.title('pred:'+str(np.where(pred1[0] == max(pred1[0]))), fontsize=22)

plt.figure('img2')
plt.imshow(img2)
plt.title('pred:'+str(np.where(pred2[0] == max(pred2[0]))), fontsize=22)
```
